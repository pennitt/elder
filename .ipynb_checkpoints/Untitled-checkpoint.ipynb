```python
import pandas as pd
import numpy as np
df = pd.read_csv('고령자 데이터 합 20221220.csv', encoding='cp949')
df = df.set_index('Unnamed: 0', drop=True)
```

```python
new_df = df.iloc[12:-1]
new_df['코로나 확진자 수'] = new_df['코로나 확진자 수'].fillna(0)
new_df
```

```python
new_df['총인구 수'] = new_df['총인구 수'].str.replace(',','').astype('int64')
for i in range(0, 9):
    new_df.iloc[:,i] = new_df.iloc[:,i].astype('int64')
    
new_df['경제성장률 (%)'] = new_df['경제성장률 (%)'].astype('int64')
new_df['코로나 확진자 수'] = new_df['코로나 확진자 수'].astype('int64')
new_df['65세 이상 사망자 수'] = new_df['65세 이상 사망자 수'].astype('int64')
```

```python
new_df["65세 이상 자진반납 현황"] = new_df["65세 이상 자진반납 현황"].fillna("0").astype(int)
new_df
```
