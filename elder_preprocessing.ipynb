```python
link = 'C:/Users/user/Desktop/고령자 민원인 예측/'
eco_growth = pd.read_csv(link + '2000_2021_65세이상_경제활동인구_연간_서울특별시.csv', encoding='cp949')
eco_act = pd.read_csv(link + '2000_2021_65세이상_경제활동참가율_연간_서울특별시.csv', encoding='cp949')
employment = pd.read_csv(link + '2000_2021_65세이상_고용율_연간_서울특별시.csv', encoding='utf-8')
not_working = pd.read_csv(link + '2000_2021_65세이상_비경제활동인구_연간_서울특별시.csv', encoding='cp949')
population = pd.read_csv(link + '2000_2021_65세이상_인구_연간_서울특별시.csv', encoding='cp949')
working = pd.read_csv(link + '2000_2021_65세이상_취업자수_연간_서울특별시.csv', encoding='cp949')
total_population = pd.read_csv(link + '201012_202112_연령별인구현황_연간_서울특별시.csv', encoding='cp949')
oil = pd.read_csv(link + '주유소_지역별_평균판매가격.csv', encoding='cp949')
returnn = pd.read_csv(link + '2015_2021_경찰청_시도 경찰청별 고령운전자 자진반납 현황_연간,전체_전국시도.csv', encoding='cp949')
disease = pd.read_csv(link + '한국산업안전보건공단_연령별질병사망자수_20211231.csv', encoding='cp949')
growth = pd.read_excel(link + '경제성장률.xls')
covid = pd.read_excel(link + '코로나바이러스감염증-19_확진환자_발생현황_221213.xlsx')
death = pd.read_csv(link + '2011_2021_연령별사망자수_(5세)_연간_서울특별시.csv')


def make_df(eng_name, kor_name):
    eng_name = pd.DataFrame(eng_name.loc[1])
    eng_name = eng_name.rename({1:kor_name}, axis=1)
    return eng_name

total_population = pd.DataFrame(total_population.iloc[0])
total_population = total_population.rename({0:'총인구 수'}, axis=1)
total_population = total_population.drop(index='행정구역', axis=1)
year = [str(i) for i in range(2010, 2022)]
total_population.index = year

oil = oil[['구분', '서울']]
oil = oil.set_index('구분')
oil_index = [i[0:4] for i in oil.index]
oil.index = oil_index
oil = oil.rename({'서울':'서울 주유비 (원)'}, axis=1)

disease = disease.set_index('구분')
disease.loc['65세 이상 질병 사망자 수', :] = disease.iloc[10:14].sum(axis=0)
disease = pd.DataFrame(disease.loc['65세 이상 질병 사망자 수'])

returnn = pd.DataFrame(returnn.iloc[0])
returnn_index = ['{} 65세 이상'.format(i) for i in range(2015, 2022)]
returnn = returnn.loc[returnn_index]
returnn_index = [i[0:4] for i in returnn.index]
returnn.index = returnn_index
returnn = returnn.rename({0:'65세 이상 자진반납 현황'}, axis=1)

growth = growth.rename(columns=growth.iloc[1])
growth.drop(growth.index[0:2])
growth = growth.iloc[2]
growth = pd.DataFrame(growth.iloc[1:])
growth_index = [str(int(i)) for i in growth.index]
growth.index = growth_index
growth = growth.rename({2:'경제성장률 (%)'}, axis=1)

covid = covid.rename(columns=covid.iloc[3])
covid = covid.drop(covid.index[0:5])
covid['일자'] = pd.to_datetime(covid['일자'])
covid_year = covid.groupby(pd.Grouper(key='일자', freq='Y'))['계(명)'].sum()
covid_year = pd.DataFrame(covid_year)
covid_year_index = [str(i.year) for i in covid_year.index]
covid_year.index = covid_year_index
covid_year = covid_year.rename({'계(명)':'코로나 확진자 수'}, axis=1)

# death
death_year = []
for i in range(2011, 2022):
    for j in range(15, 21):
        a = "{}.{}".format(i, j)
        death_year.append(a)
        
death = death[death_year][0:3]

lst = []
for i in range(len(death.iloc[2])):
    lst.append(death.iloc[2][i])

lst_num = pd.to_numeric(lst)
neww = pd.DataFrame([death.columns, lst_num])

e = 2010
for i in range(0, 65, 6):
    e += 1
    neww['{}'.format(e)] = neww.iloc[:,i:i+5].sum(axis=1)


death_year2 = [str(i) for i in range(2011, 2022)]
death2 = pd.DataFrame(neww[death_year2].iloc[1])
death2 = death2.rename({1:'65세 이상 사망자 수'}, axis=1)

eco_growth = make_df(eco_growth, '65세 이상 경제활동인구 (천 명)')
eco_act = make_df(eco_act, '65세 이상 경제활동참가율 (%)')
employment = make_df(employment, '65세 이상 고용률 (%)')
not_working = make_df(not_working, '65세 이상 비경제활동인구 (천 명)')
population = make_df(population, '65세 인구 (천 명)')
working = make_df(working, '취업자 수 (천 명)')

df_list = [eco_growth, eco_act, employment, not_working, population, working, total_population, oil, disease, returnn, growth, covid_year, death2]
df_year = pd.concat(df_list, axis=1)

df_year = df_year.sort_index()
df_year
```
